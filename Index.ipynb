```python
vars = []

for i in range(1,78):
    temp = data[data["ward"]==i]
    
    temp1 = temp['botanical_name'].value_counts().rename_axis('unique_values').reset_index(name='counts')
    var = temp1['counts'].var()
    vars.append(var)
```

```python
#variance indicates how non uniformly the ward has trees of a certain type
```
